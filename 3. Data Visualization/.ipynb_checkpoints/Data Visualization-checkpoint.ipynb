### Task 1. Detecting anomalies in car prices

1. Load the dataset from `data/vehicles_dataset_upd.csv` and display the first 5 records.


```python
df = pd.read_csv('C:\\Users\\Lietotajs\\Desktop\\ML Data Prep & Engineering\\3. Data Visualization\\data\\vehicles_dataset_upd.csv')
df.head(5)
```

2. Plot a histogram of the car price distribution, set the figure size to 8x5, and use the color `indigo`.
Don't forget to label the title and axes.

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.hist(df.price, bins=800, color='indigo')
ax.set_xlabel('Car Price')
ax.set_ylabel('Car qty')
ax.set_title('Ratio of quantity to price of the vehicle');

ax.xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
```

3. The histogram plot in its current form does not provide a clear view of the actual distribution, but it indicates the presence of outliers.
Plot a boxplot for car prices, set the figure size to 5x15, and apply formatting to display price values.
Don't forget to label the title and axes.

```python
fig, ax = plt.subplots(figsize=(5,15))
ax.boxplot(df.price)
ax.set_ylabel('Price')
ax.set_title('Ratio of quantity to price of the vehicle')
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'));
```

4. From the boxplot, we can see that outliers indeed exist, but due to their large range, the main part of the plot (the "box") is almost unnoticed.
Display descriptive statistics for car prices to examine the values representing the boundaries of the "box". Use `apply(lambda x: f"{x:0.3f}")` to format the values for better readability.

```python
df.price.describe().apply(lambda x: f"{x:0.3f}")
```

5. Output the boundaries for normal car price values using the interquartile range. Check if the minimum and maximum values fall within the interval between the obtained boundaries.

```python
def outliers_calculations(data):
    Q25= data.quantile(0.25)
    Q75= data.quantile(0.75)
    iqr = Q75 - Q25
    boundaries = (Q25 - 1.5 * iqr, Q75 + 1.5 * iqr)
    return boundaries
    
frame = outliers_calculations(df['price'])
if df.price.min()>frame[0]:
    print(f'q25 = {frame[0]}, Column "Price" minimal value = {df.price.min()}, IS included in the interval between the obtained bounds.')
else: 
     print(f'q25 = {frame[0]}, Column "Price" minimal value = {df.price.min()}, IS NOT included in the interval between the obtained bounds.')
if df.price.max()<frame[1]:
    print(f'q75 = {frame[1]}, Column "Price" maximal value = {df.price.max()}, IS included in the interval between the obtained bounds.')
else:
    print(f'q75 = {frame[1]}, Column "Price" maximal value = {df.price.max()}, IS NOT included in the interval between the obtained bounds.')
    
```

6. Calculate and output the number of outliers in the car price.

```python
out_of_frame = []
for elem in df['price']:
    if elem < frame[0]:
        out_of_frame.append(elem)
    elif elem > frame[1]:
        out_of_frame.append(elem)
print(len(out_of_frame))
```

7. Output the records where the car price falls into the outliers, sorted by price. Look at the data and consider whether these outliers may describe special price category values or other features, or if they are indeed outliers.

```python
copy = df.copy()
copy = copy[(copy['price'] <= frame[0]) | (copy['price'] >= frame[1])]
copy.sort_values(by = 'price', ascending = True)


# It seems to me that some cars are indeed outliers (especially the first two cars),
# but the Porsche 911 theoretically could have such a price, although according to 
# internet data, the most expensive version of the 911 is sold for $275,000. 
# There are also Lamborghinis and other premium-class cars in the dataset, 
# whose prices seem logical. Conclusion: cleaning is needed. 
# It's necessary to separate outliers from genuine data.
```

8. Display the last 15 records in the previously sorted dataframe with outliers to examine the features of the most expensive cars.

```python
copy.sort_values(by = 'price', ascending = False).head(15)
```

Among the cars with anomalous prices, there are indeed cars whose high price can be explained primarily by the manufacturer and low mileage. However, let's rely on the price of the main mass of cars, so remove all anomalous values and save the dataframe without anomalies into the variable df_new, then display the first 5 values.

Check the [documentation](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html) to get an understanding of its operation using usage examples.

```python
df_new = df.copy()
df_new = df_new[~(df_new['price'] > frame[1])]
df_new = df_new[~(df_new['price'] < frame[0])]
if len(df) - len(df_new) == len(copy):
    print('ok')
df_new.head(5)
```

```python
print(df_new.price.describe().apply(lambda x: f"{x:0.3f}"))
print()
```

### Task 2. Exploring the dependence of car price on the year of production

1. Formulate and display a new DataFrame containing statistics on the year of production of the car and the average value of the car price for each year (use grouping by year). Don't forget to rename the columns according to their content.

```python
grouping = df_new.groupby(['year'], as_index=False).agg({'price': 'mean'})
grouping = grouping.rename(columns={'price': 'mean_price'})
grouping

```

```python
till_21_grouping= grouping.loc[~grouping['year'].isin([2021, 2022])]
```

2. Construct a scatter plot to visualize the relationship between the car price and the year of car production. Don't forget to label the title of the graph and axes.

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(df_new.year, df_new.price, color='indigo')
ax.set_xlabel('Car Price')
ax.set_ylabel('Year of issue')
ax.set_title('Ratio year of issue to price of the vehicle');
```

3. From the graph, it can be concluded that there is an increasing trend: the newer the car, the higher its price. The exception is made by the oldest and the newest cars. Let's check if there is enough data for the outliers in terms of car production years. To do this, output the number of cars manufactured each year, using grouping by years.

```python
qty_per_year = df_new.groupby(['year'], as_index=False).agg({'price': 'count'})
qty_per_year = qty_per_year.rename(columns={'price': 'Car_qty'})
qty_per_year
```

4. Over the last two years, there is not much accumulated information — let's exclude them from the analysis. Using filtering, leave in the dataframe the cars manufactured no later than 2021, and output the first five values of the dataframe.

```python
filtered_df = qty_per_year.loc[~qty_per_year['year'].isin([2021, 2022])]
filtered_df
```

5. Repeat the actions on the filtered data to create a new dataframe containing statistics on the year of car production and the average price of cars for each year. Don't forget to rename the columns accordingly.

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(filtered_df.year, filtered_df.Car_qty, color='indigo')
ax.set_xlabel('Year of issue')
ax.set_ylabel('Car Qty')
ax.set_title('Ratio price to year of vehicle issue');
```

6. Plot a linear graph for visualizing the dependence of the car's price on the year of production [plot]. Adjust the line color to blue, the marker symbol to a diamond (D), and the marker face color to orange. Use the average car price calculated in the previous task. Don't forget to label the graph title and axes.

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(filtered_df.year, till_21_grouping.mean_price, color='blue', marker = 'D', markerfacecolor = 'orange')
ax.set_xlabel('Year of issue')
ax.set_ylabel('Mean_Car_Price')
ax.set_title('Ratio price to year of vehicle issue');
```

1. Can a conclusion be made about the existence of a linear relationship between the car's price and the year of its production? Why?
   
   Answer: The Pearson correlation coefficient, which measures the linear relationship, is close to one (0.85). A linear relationship exists. The same conclusion is supported by the graph above (except for vintage cars).
   
   

2. Can a conclusion be made about a strong correlation between the car's price and the year of its production? Why?
   
   Answer: Judging by the Pearson correlation coefficient and the clear linear relationship, there is indeed a correlation.


3. Is the existing relationship direct or inverse? Why?
   
   Your answer: The relationship is direct because the coefficient is positive.

```python
print(f'Pearson correlation coefficient:', round(filtered_df['year'].corr(till_21_grouping['mean_price']),3))
```

### Task 3. Investigating the Relationship between Price Category, Price, and Year of Manufacture

1. Plot the distribution of car prices for each price category on the same plot. Set the figure size to 12x8, transparency parameter to 0.7, and use green, orange, and red colors for low, medium, and high price categories, respectively. Don't forget to add a title to the graph, label the axes, and include a legend for the price categories.

Consider why the resulting graph looks like this.

```python
colors = {'high': 'red', 'medium':'orange', 'low':'green'}

fig, ax = plt.subplots(figsize=(12, 8))
for price_category in colors: 
    color = colors[price_category]
    data = df_new[df_new['price_category'] == price_category]
    
    ax.hist(data['price'], bins=100, color=color, alpha=0.7)
    
plt.ylabel('Frequency')
plt.xlabel('Prices') 
plt.title('Graphs of car value distribution for each of the car price categories')
plt.legend(colors, title='Price category');


# If the graph matches the request, the absence of intersections between price categories 
# is explained by the fact that they are categories meant to separate the dataset into 
# distinct parts characterized by a specific feature. Additionally, we can observe the 
# frequency of cars depending on their price. The largest number of cars belongs to 
# the medium price category, which generally corresponds to the economic classification 
# of the population in most countries.
```

2. This code creates a scatter plot of car prices by year of manufacture, with each point colored according to the price category of the car. The legend is displayed in the lower right corner of the plot.

```python
colors = {'low': 'green', 'medium': 'orange', 'high': 'red'}
fig, ax = plt.subplots(figsize=(12, 8))
for price_category in colors:
    color = colors[price_category]
    data = df_new[df_new['price_category'] == price_category]

    ax.scatter(data['price'], data['year'], c=color, label=price_category)

ax.legend(loc='lower right', title="Ценовая категория")
plt.title("The price of the vehicle and its year of manufacture")
plt.ylabel('Year of manufacture')
plt.xlabel('Price')
plt.show()
```

### Task 4: Visualizing Categorical Variables

1. Formulate and display a new dataframe containing statistics on the car manufacturer and the number of cars produced by them (use grouping). Don't forget to rename the columns according to their content.


```python
manufacturer_qty = df_new.groupby(['manufacturer'], as_index=False).agg({'price': 'count'})
manufacturer_qty  = manufacturer_qty.rename(columns={'price': 'cars_count'})
manufacturer_qty 
```

2. Build a bar chart for car manufacturers, set the figure size to 20x10, the width of the intervals (`width`) to 0.9, and the colors to alternate between `springgreen` and `dodgerblue`. Don't forget to label the title of the chart and the axes. For the X-axis, rotate the labels by 45 degrees using the `xticks` method and its `rotation` parameter.

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.bar(manufacturer_qty['manufacturer'], manufacturer_qty['cars_count'], color=['springgreen', 'dodgerblue'], width=0.9)
ax.set_xlabel('Manufacturer')
ax.set_ylabel('Count')
ax.set_title('Number of manufacturers cars')
plt.xticks(rotation=45);
```

3. Formulate and display a new dataframe containing statistics on the transmission type of the car and the number of cars with such a transmission type (use grouping). Don't forget to rename the columns according to their content.

```python
transmission = df_new.groupby(['transmission'], as_index=False).agg({'price': 'count'})
transmission = transmission.rename(columns={'price': 'cars_count'})
transmission.astype({'cars_count': 'float'})
transmission
```

4. Build a pie chart for the car transmission type, set the figure size to 12x8, format the label showing the category ratio, and set the distance from the category ratio label to the edge of the segment. Don't forget to label the graph and axes, and also display the legend separately.

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.pie(transmission['cars_count'],labels=transmission['transmission'], labeldistance=None, autopct = '%1.0f%%')
ax.set_title('Transmission type distribution');
ax.legend(loc='upper right', title = 'Transmission Type');
```
